# Task -02:  
Mini calculator: 
● Asks for two numbers (float allowed). 
● Asks which operation: +, -, *, /, ** (power), or % (modulo). 
● Performs the calculation and prints the result with 2 decimal places if it's a float. 
● If the user enters an invalid operation, print "Invalid operation – please try again next 
time."

```python
num1=float(input("Enter first number"))
num2=float(input("Enter second number"))
```

```python
print("Enter the number you want to perform\n1.addition\n2.subtraction\n3.multiplication\n4.division\n5.power\n6.modulo")
```

```python
choice=int(input())
if choice==1:
    add=num1+num2
    print(f"{add:.2f}")
elif choice==2:
    sub=num1-num2
    print(f"{sub:.2f}")
elif choice==3:
    multiply=num1*num2
    print(f"{multiply:.2f}")
elif choice==4:
    if num2==0:
        print("cannot divde by zero")
    else:
        divide=num1/num2
        print(f"{divide:.2f}")
elif choice==5:
    power=num1**num2
    print(f"{power:.2f}")
elif choice==6:
    mudulo=num1%num2
    print(f"{mudulo:.2f}")
else:
    print("Invalid operation – please try again next time.")
  
    
```
