Challenge A – Even/Odd & Multiple Checker (while loop) 
Ask user for a starting number and ending number. 
Use while loop to print every number in range. 
For each number: - If even → print "Even" - If divisible by 5 → print "Multiple of 5!" - If both → print "Even & Multiple of 5!!" 
Skip printing the number itself if it's divisible by 7 (use continue) 

```python
#user is going to enter starting ad ending number
starting_number=int(input("Enter the starting number"))
ending_number=int(input("Enter the ending number"))
```

```python
# using loop to chech rang for given problem
for i in range(starting_number,ending_number+1):
    if i%7==0:
        continue
    elif i%2==0 and i%5==0:
        print(f"{i} is Even & Multiple of 5!!")
    elif i%2==0:
        print(f"{i} is even")
    elif i%5==0:
        print(f"{i} is multiple of 5")
```

```python
#user will enter the password 
password=input("Enter your password")
```

```python
# we wll check for upper case lenght and digit 
digit=any(char.isdigit() for char in password)
upper=any(char.isupper() for char in password)
lenght=len(password)>6
```
