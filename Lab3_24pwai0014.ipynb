
Task Mini Project: Student Marks Analyzer

```python
def calculate_total_marks(marks): # first we will create a function for total marks
    return sum(marks)
```

```python
def calculate_average_marks(marks): # hen we will create function for average marks
    return sum(marks) / len(marks)
```

```python
name = input("Enter your name: ") # we wil enter the name and subject marks 
Maths = int(input("Enter maths marks: "))
English = int(input("Enter english marks: "))
Science = int(input("Enter Science marks: "))
```

```python
# then we will store the marks in a list then calculate the average and total marks
#USNG LAMDA FUNCTION WE WILL CALCULATE highest marks 
marks = [Maths, English, Science]

total_marks = calculate_total_marks(marks)

average_marks = calculate_average_marks(marks)

highest_mark = max(marks, key=lambda x: x)

print("Name:", name)
print("Marks:", marks)
print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
print("Highest Mark:", highest_mark)
```

Mini-Project: "AI Number Guessing Game Lite"

```python
import random

random_number = random.randint(1, 50)
attempts = 7
attempt = 1
```

```python
print(random_number)
```

```python
guess_num = int(input("Enter your guess: "))
while attempt <= attempts:
    if guess_num > random_number:
        print("Too high! Try lower")
        print(f"You have {attempts - attempt} attempts remaining")
        attempt += 1
        if attempt <= attempts:
            guess_num = int(input("Enter your guess: "))
    elif guess_num < random_number:
        print("Too low! Try higher")
        print(f"You have {attempts - attempt} attempts remaining")
        attempt += 1
        if attempt <= attempts:
            guess_num = int(input("Enter your guess: "))
    else:
        print(f"Congratulations! You guessed it in {attempt} tries. You're thinking like an AI optimizer!")
        break
else:
    print(f"You have lost. The secret number is {random_number}. Better luck next time")
```
