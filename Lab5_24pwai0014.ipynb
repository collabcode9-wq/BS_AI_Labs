🔐 AI Password Strength Checker Lite


```python
def check_password_strength(password):
    conditions_met = 0
    
    # Condition 1: Length at least 8
    has_number = any(char.isdigit() for char in password)
    has_uppercase = any(char.isupper() for char in password)
    has_length = len(password) >= 8
   
    if has_length:
        conditions_met += 1
    
    if has_number:
        conditions_met += 1
    
    if has_uppercase:
        conditions_met += 1
    
    if conditions_met == 0:
        feedback = "Very Weak"
    elif conditions_met == 1:
        feedback = "Weak"
    elif conditions_met == 2:
        feedback = "Moderate"
    else:
        feedback = "Strong"
    
    return feedback, conditions_met

```

```python
# Step 4: Attempts Rule
max_attempts = 5
attempts_used = 0

print(f" AI Password Strength cheker")
print(f"Create a password with: 8+ chars, 1+ number, 1+ uppercase letter")
print(f"You have {max_attempts} attempts\n")

while attempts_used < max_attempts:
    # Step 1: User Input
    password = input(f"Enter your password: ")
    attempts_used += 1
    remaining = max_attempts - attempts_used
    
    # Step 2: Check Password Strength
    strength, conditions = check_password_strength(password)
    
    # Show feedback
    print(f"Password Strength: {strength}")
    print(f"Conditions met: {conditions}/3")
    print(f"Remaining attempts: {remaining}")
    
    
    if conditions == 3:
        print(f"You created a strong password!")
        print(f"AI Training Level: Beginner → Intermediate")
        break
    
    
    if attempts_used == max_attempts and conditions != 3:
        print(f"Final Result: Password is not strong enough.")
```
