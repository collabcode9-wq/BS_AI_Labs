NumPy for Artificial Intelligence mini project

```python
import numpy as np

#  Create dataset of at least 10 students using NumPy
# Format: [Math, Science, English, Computer, Result] where Result: 1=Pass, 0=Fail
data = np.array([
    [78, 85, 80, 90, 1],
    [45, 50, 40, 55, 0],
    [92, 88, 95, 94, 1],
    [60, 65, 58, 62, 1],
    [35, 42, 38, 40, 0],
    [81, 79, 85, 83, 1],
    [55, 48, 52, 50, 0],
    [70, 72, 68, 75, 1],
    [88, 91, 89, 92, 1],
    [42, 38, 45, 40, 0],
    [95, 93, 97, 96, 1],
    [65, 60, 63, 67, 1]
])

print("Full Dataset:\n", data)

#  Separate features (X) and labels (y)
X = data[:, :4] 
y = data[:, 4] 
print("\nFeatures X:\n", X)
print("\nLabels y:", y)

#  Calculate average, maximum, and minimum marks
# Axis=0 means column-wise operation
avg_marks = np.mean(X, axis=0)
max_marks = np.max(X, axis=0)
min_marks = np.min(X, axis=0)
subjects = ['Math', 'Science', 'English', 'Computer']

print("\nSubject-wise Statistics:")
for i, subject in enumerate(subjects):
    print(f"{subject} -> Avg: {avg_marks[i]:.2f}, Max: {max_marks[i]}, Min: {min_marks[i]}")

# Find number of students who passed and failed
num_passed = np.sum(y == 1)
num_failed = np.sum(y == 0)
print(f"\nPassed: {num_passed} students")
print(f"Failed: {num_failed} students")

#  Identify students scoring above 80 and below 50
# Using total marks across all 4 subjects
total_marks = np.sum(X, axis=1)
above_80_avg = np.where(total_marks/4 > 80)[0] # indices of students
below_50_avg = np.where(total_marks/4 < 50)[0]

print(f"\nStudents with avg > 80: Index {above_80_avg}")
print(f"Students with avg < 50: Index {below_50_avg}")

# Normalize the dataset - Min-Max scaling to [0,1]
X_min = np.min(X, axis=0)
X_max = np.max(X, axis=0)
X_normalized = (X - X_min) / (X_max - X_min)
print("\nNormalized Features X:\n", np.round(X_normalized, 3))

#  Apply simple prediction logic
# Logic: If average of 4 subjects >= 60, predict Pass=1, else Fail=0
predictions = np.where(np.mean(X, axis=1) >= 60, 1, 0)
print("\nActual Results: ", y)
print("Predicted Results: ", predictions)

# Calculate accuracy of predictions
accuracy = np.mean(predictions == y) * 100
print(f"\nModel Accuracy: {accuracy:.2f}%")



```
